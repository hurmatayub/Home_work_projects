```python
def main():
  values = []

  while True:
    user_input = input("Enter a value: ")

    if user_input == "":
      break

    values.append(user_input)
  print("here's the list:", values)

main()
```

```python
def main():
    lst = []

    val = input("Enter a value: ")
    while val:
        lst.append(val)
        val = input("Enter a value: ")
    print("Here's the list:", lst)

if __name__ == '__main__':
    main()
```
