```python
import hashlib

def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

def login(email, password_to_check, stored_logins):
    hashed = hash_password(password_to_check)
    return stored_logins.get(email) == hashed


stored_logins = {
    "hurmat@example.com": hash_password("glitter123"),
    "sana@kawaii.net": hash_password("unicorns&rainbows"),
    "cooldev@code.com": hash_password("mySecretPass")
}

print(login("hurmat@example.com", "glitter123", stored_logins))
print(login("sana@kawaii.net", "wrongpass", stored_logins))
```
